# Matemática ~~diabólica~~ simbólica

```
from sympy import Symbol
```

```
x = Symbol('x')
y = Symbol('y')
```

```
type(x)
```

## Raízes

```
from sympy import solve
```

```
solve(x**2 - 4*x - 5, x)
```

```
solve(x - y + 5, x)
```

## Sistemas de equações

### de primeiro grau

```
solve([x + 5*y - 2,
       -3*x + 6*y - 15],
      [x, y])
```

### de segundo grau

```
solve([3*x - y - 9, x*y - 12], [x, y])
```

## É possível ...

### ... expandir expressões

```
((x + y)**2 * (x + 1)).expand()
```

### ... e simplificar expressões

```
from sympy import simplify
```

```
simplify(x**2 + 3*x**2 + 4*x + 2 - 8)
```

# Cálculo

### Limites

```
from sympy import limit
```

```
limit(1/x, x, 0)
```

```
limit(1/x, x, 2)
```

### Derivadas

```
from sympy import diff
```

```
diff(2*x**2 + 3*x + 2)
```

```
diff(2*x**2 + 3*x + 2, x)
```

```
diff(2*x**2 + 3*x + 2, x, x)
```

```
diff(2*x**2 + 3*x + 2, x, x, x)
```

### Inclinação da reta no ponto

```
diff(x ** 2, x).subs(x, 2).evalf()
```

### Integral indefinida

```
from sympy import integrate
```

```
integrate(x**2 + 1, x)
```

### Integral definida

```
integrate(x**2 + 1, (x, 0, 2))
```
