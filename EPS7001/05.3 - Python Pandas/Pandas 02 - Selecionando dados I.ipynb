# Selecionando dados

Vamos usar um novo conjunto de dados aqui para demonstrar como lidar com conjuntos de dados maiores. Este é um subconjunto das 311 solicitações de serviço do NYC Open Data.

````python
!wget https://raw.githubusercontent.com/jvns/pandas-cookbook/master/data/311-service-requests.csv
````

````python
import pandas as pd
````

````python
complaints = pd.read_csv('311-service-requests.csv')
````

````python
type(complaints)
````

Dependendo da versão do seu Pandas, você pode ver um erro como "DtypeWarning: Columns (8) have mixed types".

Isso significa que encontrou um problema ao ler nossos dados. Nesse caso, quase certamente significa que há colunas nas quais algumas das entradas são strings e algumas são inteiros.

Por enquanto, vamos ignorá-lo e esperar que não nos deparemos com um problema, mas, a longo prazo, precisaríamos investigar esse aviso.

# O que há no ```DataFrame```? (o sumário)

Quando você imprime um grande ```DataFrame```, ele mostra apenas as primeiras linhas.


````python
complaints
````

# Selecionando colunas e linhas

Para selecionar uma coluna, indexamos com o nome da coluna, assim:

````python
complaints['Complaint Type']
````

Para obter as primeiras 5 linhas de um ```DataFrame```, podemos usar um slice: ```df[:5]```.

Essa é uma ótima maneira de ter uma ideia de que tipo de informação está no ```DataFrame```. Reserve um minuto para analisar o conteúdo e ter uma ideia desse conjunto de dados.

````python
complaints[:5]
````

Podemos combiná-los para obter as primeiras 5 linhas de uma coluna:

````python
complaints['Complaint Type'][:5]
````

e não importa em que direção fazemos isso:

````python
complaints[:5]['Complaint Type']
````

# Selecionando várias colunas

E se quisermos apenas saber o tipo de reclamação e o bairro, mas não o resto das informações? O Pandas facilita muito a seleção de um subconjunto das colunas: apenas indexe com a lista de colunas que você deseja.

````python
complaints[['Complaint Type', 'Borough']]
````

Isso nos mostrou um resumo e podemos ver as primeiras 10 linhas:

````python
complaints[['Complaint Type', 'Borough']][:10]
````

# Qual é o tipo de reclamação mais comum?

Esta é uma pergunta muito fácil de responder! Existe um método ```.value_counts()``` que podemos usar:

````python
complaints['Complaint Type'].value_counts()
````

Se quiséssemos apenas as 10 queixas mais comuns, podemos fazer isso:

````python
complaint_counts = complaints['Complaint Type'].value_counts()
complaint_counts[:10]
````

Mas fica ainda melhor! Nós podemos plotar!

````python
complaint_counts[:10].plot(kind='bar')
````
