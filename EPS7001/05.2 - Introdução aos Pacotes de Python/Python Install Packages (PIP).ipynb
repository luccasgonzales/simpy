# Pacotes

Pacotes trazem funcionalidades adicionais aos recursos nativos de Python.

Por exemplo, o pacote ```math``` abaixo

````
sum([1,2,3])
````

````
from math import hypot
````

````
hypot(1, 1.272)  
````

````
# numbers from https://en.wikipedia.org/wiki/Kepler_triangle
print(hypot(1.618, 2.058)/hypot(1.272, 1.618))
print(hypot(1.272, 1.618)/hypot(1, 1.272))
````

````
import math
````

````
math.hypot(1, 1.272)
````

````
math.sqrt(1**2 + 1.272**2)
````

````
math.factorial(5)
````

````
math.log(math.exp(1))
````

Existem pacotes **nativos** (como o ```math``` acima) que não precisam ser instalados, apenas importados.

Existem ainda pacotes que não vêm instalados no Python por padrão. Este tipo de pacote precisa ser instalado antes de ser importado. Caso contrário veremos um ```ModuleNotFoundError``` como abaixo.

````
import simpy
````

# PIP

Para instalar pacotes não nativos no Python, existe o programa ```pip```, que serve exatamente para esse propósito.

Note que o programa ```pip``` não é executado por dentro do Python, mas sim em um terminal.

Em um ambiente de Jupyter Notebooks como o nosso, podemos rodar comandos de terminal nas próprias células do notebook. Para isso, apenas devemos começar a instrução com um sinal de exclamação (!)

````
!pip3 install simpy
````

````
import simpy
````

O programa ```pip``` ainda oferece um comando adicional para sabermos qual a versão de um pacote que temos instalado (e se temos) em nosso Python. Esse é o comando ```freeze```.

````
!pip3 freeze | grep simpy
````
